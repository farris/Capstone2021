```python
from PIL import Image
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
import torch
from skimage import io
import numpy as np
import os
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/CDS Capstone Project/Data/Monkey Data.xlsx")
```

```python
monkey_data = df[df['tiff present'] == True]
monkey_data = monkey_data.iloc[1:,:]
```

```python
monkey_data.shape
```

```python
data_dir = '/content/drive/MyDrive/CDS Capstone Project/Data/Raw'
dir = monkey_data['original scan folder (tiff)'].unique()
# print(dir)
for path in dir:
  sub_path = os.path.join(data_dir, path)
  sub_path = sub_path.replace('\'','')
  
  for file_name in os.listdir(sub_path):
    if(file_name.endswith('.tif')):
      image_path = os.path.join(sub_path, file_name)
      # print(image_path)
      img = io.imread(image_path)
      img = torch.from_numpy(img.astype(np.float32))
      print(image_path, img.shape)

```
