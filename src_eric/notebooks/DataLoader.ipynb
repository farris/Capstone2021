```python
class ImageData(Dataset):
  def __init__(self):
    self.features = []
    data_dir = 'gdrive/My Drive/CDS Capstone Project/Data/Raw'
    dir = monkey_dir_1.unique()
    # print(dir)
    for path in dir:
      sub_path = os.path.join(data_dir, path)
      sub_path = sub_path.replace('\'','')
      
      for file_name in os.listdir(sub_path):
        if(file_name.endswith('.tif')):
          image_path = os.path.join(sub_path, file_name)
          # print(image_path)
          
    def __len__(self):
      return len(self.features)
    def __getitem__(self,idx):
      img = io.imread(image_path)
      img = torch.from_numpy(img.astype(np.float32))
      self.features.append(img)
      return self.features[idx]


```

```python
dataset = ImageData()
```

```python
loader = torch.utils.data.DataLoader(dataset, batch_size = 1, num_workers = 2)
```

```python
import numpy as np

def get_random_crop(image, crop_width,crop_height, crop_depth):

    max_x = image.shape[0] - crop_width
    max_y = image.shape[1] - crop_height
    max_z = image.shape[2] - crop_depth

    x = np.random.randint(0, max_x)
    y = np.random.randint(0, max_y)
    z = np.random.randint(0, max_z)

    crop = image[x: x + crop_width, y: y + crop_height,z: z+crop_depth]

    return crop

boct.vol[0] = np.random.randint(0, 256, (512,512,2048))
random_crop = get_random_crop(boct.vol[0], 143, 228, 194)
```
